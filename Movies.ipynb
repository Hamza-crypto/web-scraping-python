```python
BASE_URL = 'https://www.moviequotedb.com'

IMAGES_FOLDER_NAME = "Movies_images/"
Explaination = ""
counter = 0
items = {}

import requests
import urllib.request
import pandas as pd
from bs4 import BeautifulSoup
for i in range(5):
    print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$  "+str(i))
    res = requests.get(BASE_URL)
    soup = BeautifulSoup(res.content, 'html.parser')


    movies = soup.find('div',{'class':'home_content'})
    movies = movies.find_all('a')


    for i in movies:
#         if counter>2: break
#         counter+=1    
        title = i.img['title']
        print(title)
        image_url = i.img['src']
        print(image_url)
        urllib.request.urlretrieve(image_url,IMAGES_FOLDER_NAME+title+".jpg")
        movie_link = BASE_URL +i['href']
        print(movie_link)

        res = requests.get(movie_link)
        soup = BeautifulSoup(res.content, 'html.parser')
        quotes = soup.find_all('div',{'class':'padded'})
        for q in quotes:
            try:
                author = q.find('b')
                author = author.text
                print(author)
                quote_text = q.text
                quote_text = quote_text.replace(author,'')
                print(quote_text)
                print('----------------')
                items[counter] = [ quote_text.replace(": ","").replace("[",""), ' " '+author+' " ', ' " '+title+' " ', '"Movies"', ' " '+IMAGES_FOLDER_NAME+title+".jpg"+' " ' , Explaination ]
                counter+=1
            except:
                continue
                print("**************************")

        print("-------------------")

df_links = pd.DataFrame.from_dict(items, orient='index', columns=["Quote Text" , "Author" , "Movie Name" , "Category" , "Image path", "Description"] )
# df_links.to_csv("output.csv",  index = False)
df_links.to_excel(r'output.xlsx',  index = False)
print("Successfully Completed | Ceated excel file")
```
